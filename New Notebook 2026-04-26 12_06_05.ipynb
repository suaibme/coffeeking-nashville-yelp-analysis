```sql
-- Finding the total count of shops/businesses from the Yelp data (business table)
SELECT COUNT(*) AS total_businesses
FROM yelp_academic_dataset_nash_business;

-- Calculating mean, max, standard deviation, media, and min for the stars table
SELECT
ROUND(AVG(stars), 2)    AS mean_stars,
PERCENTILE(stars, 0.5)  AS median_stars,
MIN(stars)              AS min_stars,
MAX(stars)              AS max_stars,
ROUND(STD(stars), 2)    AS std_stars
FROM yelp_academic_dataset_nash_business;

-- Review count - mean, max, median, min
SELECT
  ROUND(AVG(review_count), 2)   AS mean_reviews,
  PERCENTILE(review_count, 0.5) AS median_reviews,
  MIN(review_count)             AS min_reviews,
  MAX(review_count)             AS max_reviews
FROM yelp_academic_dataset_nash_business;

-- rating distribution: how many businesses fall in each star bracket
SELECT
  stars,
  COUNT(*) AS total_businesses_understars
FROM yelp_academic_dataset_nash_business
GROUP BY stars
ORDER BY stars;

-- what are the top ten businesses by review
SELECT
  name,
  stars,
  review_count,
  is_open
FROM yelp_academic_dataset_nash_business
ORDER BY review_count DESC
LIMIT 10;

-- Top 10 most reviewed but lowest rated, the underperformers
SELECT
  name,
  stars,
  review_count,
  is_open
FROM yelp_academic_dataset_nash_business
ORDER BY review_count DESC, stars ASC
LIMIT 10;
```
